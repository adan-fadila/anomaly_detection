```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import RobustScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# Load and preprocess data
data = pd.read_csv('yahoo_sub_5.csv')
data.fillna(method='ffill', inplace=True)  # Handle missing values

# Normalize data
scaler = RobustScaler()
data['value_scaled'] = scaler.fit_transform(data[['value_1']])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 7))
plt.plot(data.index, data['value_scaled'], label='Time Series Data')  # Replace 'value' with the column name
plt.title('Time Series Data')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

decomposition = seasonal_decompose(data['value_scaled'], model='additive', period=24)  # Adjust 'period' to your data frequency
decomposition.plot()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(data['value_scaled'], bins=30, edgecolor='k')
plt.title('Value Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Create sequences for LSTM
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Dense, BatchNormalization
from tensorflow.keras import regularizers
from keras.callbacks import EarlyStopping
from sklearn.model_selection import train_test_split

def create_sequences(data, time_steps=30):
    sequences, targets = [], []
    for i in range(len(data) - time_steps):
        sequences.append(data[i:i + time_steps])
        targets.append(data[i + time_steps])
    return np.array(sequences), np.array(targets)

time_steps = 30
X, y = create_sequences(data['value_scaled'].values, time_steps)

Split into training and testing sets
split_idx = int(0.8 * len(X))
X_train, X_test = X[:split_idx], X[split_idx:]
y_train, y_test = y[:split_idx], y[split_idx:]
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
model = Sequential([
    LSTM(64, activation='relu', return_sequences=True, kernel_regularizer=regularizers.l2(0.01), input_shape=(30, 1)),
    Dropout(0.5),
    LSTM(32, activation='relu', return_sequences=True, kernel_regularizer=regularizers.l2(0.01)),
    Dropout(0.5),
    LSTM(16, activation='relu'),
    Dropout(0.5),
    Dense(1)
])

early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

model.compile(optimizer='adam', loss='mean_squared_error')
model.summary()
# Train model
model.fit(X_train, y_train, epochs=50, batch_size=64, validation_data=(X_val, y_val), callbacks=[early_stopping])
# Predict on test data
X_test_pred = model.predict(X_test)
```

```python
# Assuming 'y_test' corresponds to the test set and its size matches 'reconstruction_error'
reconstruction_error = np.abs(y_test - X_test_pred.flatten())  # Absolute difference between actual and predicted values

# Define a threshold for anomaly detection (can be set manually or using statistical methods)
threshold = np.mean(reconstruction_error) + 2 * np.std(reconstruction_error)
test_indices = data.index[-len(reconstruction_error):]  # Get indices for the test set
data['is_anomaly'] = 0  # Initialize all as non-anomalous
data.loc[test_indices, 'is_anomaly'] = [
    1 if error > threshold else 0 for error in reconstruction_error
]

# Visualize anomalies
plt.figure(figsize=(14, 7))
plt.plot(data.index, data['value_scaled'], label='Time Series Data')
plt.scatter(
    data[data['is_anomaly'] == 1].index, 
    data[data['is_anomaly'] == 1]['value_scaled'], 
    color='red', label='Anomalies'
)
plt.title('LSTM-based Anomaly Detection')
plt.legend()
plt.show()
```

```python
# Assign anomaly column only to the test set indices
test_indices = data.index[-len(reconstruction_error):]  # Get the test indices
data.loc[test_indices, 'is_anomaly'] = [
    1 if error > threshold else 0 for error in reconstruction_error
]

# Extract the portion of the dataset used for testing
test_data = data.iloc[-len(y_test):]

# Visualize anomalies
plt.figure(figsize=(14, 7))
sns.lineplot(
    x=test_data.index,
    y=scaler.inverse_transform(y_test.reshape(-1, 1)).flatten(),
    label='Original Data'
)

# Extract anomalies for visualization
anomalies = test_data[test_data['is_anomaly'] == 1]
sns.scatterplot(
    x=anomalies.index,
    y=scaler.inverse_transform(y_test.reshape(-1, 1)).flatten()[anomalies.index - test_indices[0]],
    color='red',
    label='Anomalies'
)

plt.title('LSTM-based Anomaly Detection')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'X_test_pred' contains the LSTM predictions for the test set
# and 'y_test' contains the true values for the test set.

# Context window size
context_window = 10

# Compute contextual reconstruction error
contextual_reconstruction_error = []
for i in range(len(y_test) - context_window + 1):
    true_context = y_test[i:i + context_window]
    pred_context = X_test_pred.flatten()[i:i + context_window]
    error = np.mean(np.abs(true_context - pred_context))  # Average error over context window
    contextual_reconstruction_error.append(error)

# Define a threshold for context anomaly detection
threshold_context = np.mean(contextual_reconstruction_error) + 1.5* np.std(contextual_reconstruction_error)

# Extend contextual reconstruction error to match the original dataset length (pad with NaNs)
contextual_reconstruction_error = [np.nan] * (context_window - 1) + contextual_reconstruction_error

# Mark anomalies
test_indices = data.index[-len(contextual_reconstruction_error):]
data['is_context_anomaly'] = 0  # Initialize all as non-anomalous
data.loc[test_indices, 'is_context_anomaly'] = [
    1 if error > threshold_context else 0 for error in contextual_reconstruction_error
]

# Visualize anomalies
plt.figure(figsize=(14, 7))
plt.plot(data.index, data['value_scaled'], label='Time Series Data')
plt.scatter(
    data[data['is_context_anomaly'] == 1].index,
    data[data['is_context_anomaly'] == 1]['value_scaled'],
    color='red', label='Contextual Anomalies'
)
plt.title('LSTM-based Contextual Anomaly Detection')
plt.legend()
plt.show()
```

```python
# Extract test data
test_data = data.loc[test_indices]

# Visualize test data with contextual anomalies
plt.figure(figsize=(14, 7))
plt.plot(
    test_data.index, test_data['value_scaled'], label='Test Time Series Data'
)
plt.scatter(
    test_data[test_data['is_context_anomaly'] == 1].index,
    test_data[test_data['is_context_anomaly'] == 1]['value_scaled'],
    color='red', label='Contextual Anomalies'
)
plt.title('LSTM-based Contextual Anomaly Detection (Test Data)')
plt.xlabel('Time')
plt.ylabel('Value (Scaled)')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Collective anomaly detection parameters
window_size = 10  # Size of the collective anomaly window

# Compute collective reconstruction error
collective_reconstruction_error = []
for i in range(len(y_test) - window_size + 1):
    true_window = y_test[i:i + window_size]
    pred_window = X_test_pred.flatten()[i:i + window_size]
    error = np.mean(np.abs(true_window - pred_window))  # Average error over the window
    collective_reconstruction_error.append(error)

# Define a threshold for collective anomaly detection
threshold_collective = np.mean(collective_reconstruction_error) + 2 * np.std(collective_reconstruction_error)

# Extend reconstruction error to match test data length (pad with NaNs)
collective_reconstruction_error = [np.nan] * (window_size - 1) + collective_reconstruction_error

# Mark collective anomalies in test data
test_indices = data.index[-len(collective_reconstruction_error):]
test_data = data.iloc[-len(collective_reconstruction_error):].copy()  # Extract test data
test_data['is_collective_anomaly'] = [
    1 if error > threshold_collective else 0 for error in collective_reconstruction_error
]

# Visualize test data with collective anomalies
plt.figure(figsize=(14, 7))
plt.plot(test_data.index, test_data['value_scaled'], label='Test Data', color='blue')

# Highlight collective anomalies (entire windows as a group)
for i in range(len(test_data) - window_size + 1):
    if test_data['is_collective_anomaly'].iloc[i]:
        plt.plot(
            test_data.index[i:i + window_size],
            test_data['value_scaled'].iloc[i:i + window_size],
            color='red', linewidth=2.5, label='Collective Anomaly' if i == 0 else "", zorder=5
        )

plt.title('LSTM-based Collective Anomaly Detection (Test Data Only)')
plt.xlabel('Time')
plt.ylabel('Scaled Value')
plt.legend()
plt.show()
```
